```python
import pygmt as gmt
import numpy as np
import pandas as pd
import os 
import matplotlib.pyplot as plt
import xarray as xr
%matplotlib inline
```

```python
df=pd.read_csv('L:/codes_paper1/figure4/data_figure4/Global_temp_anomaly_1850_2099.txt',sep='\t',header=None).to_numpy()
thyr=[2015,2028,2038,2049,2057,2066,2073]
than=np.arange(0.5,4.0,0.5)
```

```python
fig, ax = plt.subplots(figsize=(6, 4.5))
ax.plot(df[:,0], df[:,-1],'b',linewidth=2.5)

ax.set(ylabel="Global Mean Temperature Anomaly (\N{DEGREE SIGN}C)\n w.r.t. observation (1982- 2019)", xlabel='Years',
       title='',ylim=[-2,5.5],xlim=[1850, 2100])
ax.grid(color='k', linestyle='-', linewidth=0.4, alpha=0.25)

for i in range(7):
    ax.plot((1850,thyr[i]),(than[i],than[i]),'k')
    ax.plot((thyr[i], thyr[i]),(than[i],-2),'k')
fig.savefig("L:/codes_paper1/figure4/figures/GMTA_TSplot_using_matplotlib.png")
plt.show()
```
